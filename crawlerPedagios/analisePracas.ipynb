```python
df = pd.read_csv('pracas.csv', sep=';', encoding='ISO-8859-1')
```

```python
latLong = df[['latitude']].apply(lambda row: f"{row['latitude']},{row['longitude']}", axis=1).tolist()

print(latLong)
```

```python
long = df[['longitude']].apply(lambda row: f"{row['longitude']}", axis=1).tolist()

print(long)
```

```python

lat = [local.replace(',', '.') for local in lat]
print(lat)
```

```python
long = [local.replace(',', '.') for local in long]
print(long)
```

```python
latLong = []
for numLat, numLong in zip(lat, long):
    novoNum = f'{numLat},{numLong}'
    latLong.append(novoNum)
print(latLong)
```
