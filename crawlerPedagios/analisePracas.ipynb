## Criando uma lista de latitudes e longitudes para realizar requisições na API "tomtom"

```python
latLong = df[['latitude', 'longitude']]
```

```python
latLong
```

```python
long = df[['longitude']].apply(lambda row: f"{row['longitude']}", axis=1).tolist()

print(long)
```

```python
lat = df[['latitude']].apply(lambda row: f"{row['latitude']}", axis=1).tolist()

print(lat)
```

```python

lat = [local.replace(',', '.') for local in lat]
print(lat)
```

```python
long = [local.replace(',', '.') for local in long]
print(long)
```

```python
latLong = []
for numLat, numLong in zip(lat, long):
    novoNum = f'{numLat},{numLong}'
    latLong.append(novoNum)
print(latLong)
```

## Criando uma lista de cidades para realizar requisições na api "ClimaTempo" 

```python
listaMunicipios = df['municipio'].unique()
```

```python
print(listaMunicipios)
```

```python
listaMunicipios = listaMunicipios.tolist()
```

```python
print(len(listaMunicipios))
```

```python
print(listaMunicipios)
```

```python
#6=Pesquisa ID da Cidade
listaIDCidades = [] 
cont = 1
for cidade in listaMunicipios:
    if cont % 10 == 0:
        time.sleep(70)
    iCITY = cidade
    iURL = "http://apiadvisor.climatempo.com.br/api/v1/locale/city?name=" + iCITY + "&token=" + iTOKEN
    iRESPONSE = requests.request("GET", iURL)
    iRETORNO_REQ = json.loads(iRESPONSE.text)
    print(iRETORNO_REQ)
    listaIDCidades.append(iRETORNO_REQ)
    print(cont)
    cont += 1
 
```

```python
print(listaIDCidades)
```

```python
backupID = listaIDCidades
```

```python
print(type(listaIDCidades))
```

```python
dados = listaIDCidades
```

```python
dados_achatados = [item for sublist in dados for item in sublist]

# Converter em DataFrame
df = pd.DataFrame(dados_achatados)

# Exibir o DataFrame
print(df)
```

```python
df.to_csv('idCidades.csv',index=False)
```
